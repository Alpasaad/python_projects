# Histograms are a great way to visualize individual color components
# here we are gonna to implement histogram of each color 

```python
import cv2
import numpy as np

# We need to import matplotlib to create our histogram plots
from matplotlib import pyplot as plt
```

```python
image=cv2.imread("news.jpg")
plt.hist(image.ravel(), 256, [0, 256]);
plt.show()
```

```python
color=('b','g','r')
for i,col in enumerate(color):
    histogram=cv2.calcHist(image,[i],None,[256],[0,255])
    plt.plot(histogram,color=col)
    plt.xlim([0,256])

plt.show()
```
